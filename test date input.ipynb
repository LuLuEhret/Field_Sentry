```python
import re
from datetime import datetime, timedelta
from pytz import timezone
from dateutil import parser
from insolAPI.WebAPI import API
import pandas as pd
import pendulum as pdl
```

```python

def time_args_definition(start_date, stop_date):
    time_args = dict(
    start=str(start_date),
    stop=str(stop_date),
    timezone=timezone('Europe/Zurich'),
    )
    return time_args

def download_logs(**time_args):
    api = API("lucas.ehret@insolight.ch", "toupie02Ins", dev_space=False)
    df_test = api.api.get_sensor_channels_logs_joined(**time_args, sensor_type=api.SensorsTypes.PAR, install=25)
    return df_test

def validate_date(date_str):
    #1st split the date string into year, month and day
    #convert date_parts and time_parts into integers
    try:
        res = bool(parser.parse(date_str))
    except ValueError:
        res = False
    return res



# Prompt user to enter a date
date_input = input("Enter a date (yyyy-mm-dd HH:MM): ")

# Validate the date format
if validate_date(date_input):
    #convert date_input into datetime object
    date_obj = parser.parse(date_input)
    if date_obj <= datetime.now():
        if date_obj.year >= 2022:
            print(f"Selected date is: {date_obj}")
            # add 30min  to the date
            new_date = date_obj + timedelta(days= 0,hours=12, minutes=30)
            print(f"New date is: {new_date}")
            time_args = time_args_definition(date_obj, new_date)
            # api = API("lucas.ehret@insolight.ch", "toupie02Ins", dev_space=False)
            # df = api.get_sensor_channels_logs_joined(**time_args, sensor_type=api.SensorsTypes.PAR, install=25)
            # df.head(5)
        else:
            print("The year should be greater than 2022")
    else:
        print("The date is in the future")
else:
    print("Invalid date format. Please enter date in yyyy-mm-dd HH:MM format.")
```
